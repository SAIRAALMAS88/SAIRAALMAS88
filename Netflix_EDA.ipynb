```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/Netflix_dataset.csv')
```

2. How many rows and columns does the dataset contain?

```python

rows = len(df)
cols = len(df.columns)

print(f"Number of rows: {rows}")
print(f"Number of columns: {cols}")
```

3. How many unique show types (type) are present in the dataset?

```python

unique_show_types = df['type'].nunique()
print(f"Number of unique show types: {unique_show_types}")
```

4. Count the number of shows released per year (release_year).

```python

shows_per_year = df.groupby('release_year')['show_id'].count()

shows_per_year
```

5. Which year had the most releases?

```python

year_with_most_releases = shows_per_year.idxmax()
print(f"The year with the most releases: {year_with_most_releases}")
```

6. Filter the dataset to show only the TV shows that were added in 2021.

```python

tv_shows_added_in_2021 = df[(df['type'] == 'TV Show') & (df['release_year'] == 2021)]
tv_shows_added_in_2021
```

7. Find all the shows directed by "Kirsten Johnson."

```python

shows_directed_by_kirsten_johnson = df[df['director'] == 'Kirsten Johnson']
shows_directed_by_kirsten_johnson
```

8. Identify all shows that were produced in the United States and have a rating of
"TV-MA."

```python

shows_usa_tvma = df[(df['country'] == 'United States') & (df['rating'] == 'TV-MA')]
shows_usa_tvma
```

9. List all shows where "Samuel Jouy" is part of the cast.

```python

shows_with_samuel_jouy = df[df['cast'].str.contains('Samuel Jouy', na=False)]
shows_with_samuel_jouy
```

10.Find all movies with a duration of more than 90 minutes (ensure the duration column is parsed correctly for numerical comparisons).

```python

# Convert duration to numeric, handling non-numeric values
df['duration_numeric'] = pd.to_numeric(df['duration'].str.extract('(\d+)')[0], errors='coerce')

# Filter for movies with duration > 90
movies_over_90_minutes = df[(df['type'] == 'Movie') & (df['duration_numeric'] > 90)]

movies_over_90_minutes
```

11. Find all shows that belong to both "Crime TV Shows" and "International TV Shows."

```python

crime_international_shows = df[df['listed_in'].str.contains('Crime TV Shows', na=False) & df['listed_in'].str.contains('International TV Shows', na=False)]
crime_international_shows
```

12.Find all TV shows added after January 1, 2020, that were released in India.

```python

tv_shows_india_after_2020 = df[(df['date_added'] > '2020-01-01') & (df['country'] == 'India') & (df['type'] == 'TV Show')]

tv_shows_india_after_2020
```

13.Sort the dataset by date_added in descending order.

```python

df_sorted_by_date = df.sort_values(by='date_added', ascending=False)

df_sorted_by_date
```

14.What is the most recent addition?

```python

most_recent_addition = df_sorted_by_date.iloc[0]

most_recent_addition
```

```python

average_release_year_by_rating = df.groupby('rating')['release_year'].mean()
average_release_year_by_rating
```

20.Which genre category (listed_in) appears most frequently in the dataset?

```python

genre_counts = df['listed_in'].str.split(', ').explode().value_counts()
most_frequent_genre = genre_counts.index[0]
print(f"The most frequent genre is: {most_frequent_genre}")
```
